```python
from re import I
import numpy as np 
# imprimir las opciones de los metodos
print("1:Metodo de biseccion")
print("2:Metodo de Newton")
print("3:Metodo de la secante")
# escojer el metodo a utilizar
metodo=(int(input("Elija el metodo a usar (1,2 o 3): ")))
# imprimir los ejercicios a escojer
print("1: x**3-2*x**2-5, [1,4]")
print("2: x**3+3*x**2-1, [-3,-2]")
print("3: x-cos(x), [0,pi/2]")
print("4: x-0.8-0.2*sen(x), [0,pi/2]")
# escojer el ejercicio
ejercicio=(int(input("elija el ejercicio a resolver (1,2,3 o 4): ")))
 # definir la funcion dependiendo del numero del ejercicio escojido
def f(x,ejercicio):
    if ejercicio==1:
      return x**3-2*x**2-5
    elif ejercicio==2:
      return x**3+3*x**2-1
    elif ejercicio==3:
      return x- np.cos(x)
    elif ejercicio==4:
      return x-0.8-0.2*np.sin(x)
    else:
      print("error vuelva a digitar el numero del ejercicio")
  # derivada de las funciones dependiendo del numero escojido
def dervf(x,ejercicio):
      if ejercicio==1:
        return 3*x**2-4*x
      elif ejercicio==2:
        return 3*x**2+6*x
      elif ejercicio==3:
        return 1+np.sin(x)
      elif ejercicio==4:
        return 1-0.2*np.cos(x)
  # definimos el metodo de biseccion
def bisec(a,b,tol,ejercicio):
    if f(a,ejercicio) * f(b,ejercicio) > 0:
        return print("No hay raiz en ese intervalo")
    else:
        p=(a + b) / 2
        maxiter =1000
        iter=0
        while np.abs(f(p,ejercicio)) > tol and iter < maxiter:
            iter +=1
            if np.abs(f(a,ejercicio)) <= tol:
                return a
            elif np.abs(f(b,ejercicio)) <= tol:
                return b 
            elif f(a,ejercicio) * f(p,ejercicio) < 0:
                b = p
            else:
                a = p
            p = (a + b) / 2  
        return "Numero de interaciones:",iter,"Raiz:", p
  # definimos el metodo de newton
def newt(p,ejercicio):
    iter = 0
    tol=0.0001
    maxiter=1000
    if f(p,ejercicio)==0:
      h=p
    else:
        h= p - (f(p,ejercicio) / dervf(p,ejercicio))
        while np.abs(f(p,ejercicio)) >= tol and iter < maxiter:
            p = h
            h= p - (f(p,ejercicio) / dervf(p,ejercicio))
            iter += 1
    return "Numero de interaciones:",iter,"Raiz:", h
  # definimos el metodo de la secante
def sec(p0,p1,ejercicio):
    iter = 0
    tol=0.0001
    maxiter=1000
    if f(p0,ejercicio)==0:
      h=p0
    elif f(p1,ejercicio)==0:
      h=p1
    else:
        h= p1 - ((f(p1,ejercicio)*(p1-p0)) / (f(p1,ejercicio)-f(p0,ejercicio)))
        while np.abs(f(p1,ejercicio)) >= tol and iter < maxiter:
            p0=p1
            p1 = h
            h= p1 - ((f(p1,ejercicio)*(p1-p0)) / (f(p1,ejercicio)-f(p0,ejercicio)))
            iter += 1
    return 'Numero de interaciones:',iter,'Raiz:', h
  # definimos los intervalos para poder utilizar el metodo de biseccion
if metodo == 1:
    if ejercicio == 1:
        a, b = 1, 4
    elif ejercicio == 2:
        a, b = -3, -2
    elif ejercicio == 3:
        a, b = 0, np.pi/2
    elif ejercicio == 4:
        a, b = 0, np.pi/2
    print(bisec(a, b, tol, ejercicio))
  # definimos un valor inicial para implementar el metodo de newton
elif metodo == 2:
    p = float(input("Ingrese el valor inicial que pertenezca al intervalo: "))
    print(newt(p, ejercicio))
  # definimos dos puntos para implementar el metodo de la secante
elif metodo==3:
    p0=float(input("Digita valor inicial que pertenezca al intervalo: "))
    p1=float(input("Digita un segundo valor que pertenezca al intervalo: "))
    print(sec(p0,p1,ejercicio))
else:
  print("error vuelva a digitar el numero del metodo")
```
